```python
def GenderCheck():
    while 1:
        gender = input("Enter Your Gender: Male, Female, Other").lower()
        if gender == "other":
            break
        elif gender == "female":
            print("You can Marry a boy")
        elif gender == "male":
            print("You can Marry a girl")
        else:
            print("You are beautiful")


GenderCheck()


```

```python
def GenderCheck(gender):
    if gender == "other":
        print("Your choice")
    elif gender == "female":
        print("You can Marry a boy")
        GenderCheck(input("Enter Your Gender: Male, Female, Other").lower())
    elif gender == "male":
        print("You can Marry a girl")
        GenderCheck(input("Enter Your Gender: Male, Female, Other").lower())
    else:
        print("You are beautiful guy")
        GenderCheck(input("Enter Your Gender: Male, Female, Other").lower())


GenderCheck(input("Enter Your Gender: Male, Female, Other").lower())


```
