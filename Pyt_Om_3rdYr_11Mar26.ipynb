```python
scores = [78, 40, 92, 34, 67, 88, 53, 41, 76, 29, 64, 81, 39, 55, 72, 90, 47, 68, 33, 84, 0, -5, 100, 40, 75]

print("List of scores:", scores)
print("List containing the first 10 scores:", scores[0:10])  
print("List containing the last 5 scores:", scores[-5:])
print("A list containing every second score from the original list:", scores[0:25:2])

for score in scores:
    if score >= 40:
        print(score, "is Passed")
    else:
        print(score, "is Failed")


# Use list comprehension to create:
# A list of all passing scores.
# A list of all scores increased by 5 marks (bonus marks).
# A list containing only the scores greater than 75.
#5. Display:
# The number of students who passed.
# The number of students who failed.

# passing_scores = [x for x in scores if x >= 40]
passing_scores = [score for score in scores if score >= 40]
print("List of passing scores:", passing_scores)
bonus_scores = [score + 5 for score in scores]
print("List of scores increased by 5 marks (bonus marks):", bonus_scores)
high_scores = [score for score in scores if score > 75]
print("List of scores greater than 75:", high_scores)   
num_passed = len(passing_scores)
num_failed = len(scores) - num_passed
print("Number of students who passed:", num_passed)
print("Number of students who failed:", num_failed)
```
