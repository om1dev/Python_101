```python
try:
    amount = float(input("Enter Amount: "))

    if amount <= 0:
        print("Amount must be greater than 0")

    elif amount <= 1000:
        discount = 5
    elif amount <= 10000:
        discount = 10
    elif amount <= 20000:
        discount = 15
    else:
        discount = 20

    discount_amount = amount * (discount / 100)
    final_amount = amount - discount_amount

    print("\n----- BILL SUMMARY -----")
    print(f"Your Amount is Rs. {amount:.2f}")
    print(f"Your Discount is {discount}%")
    print(f"Your Discount Amount is Rs. {discount_amount:.2f}")
    print(f"Your Final Amount is Rs. {final_amount:.2f}")

except ValueError:
    print("Please enter a valid numeric amount")
```
