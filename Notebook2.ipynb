```python
amount = int(input("Enter Amount: "))
if amount <= 1000:
    discount = 5
elif amount <= 10000:
    discount = 10
elif amount >= 20000:
    discount = 15
else:
    discount = 10

print(f"Your Amount is Rs. {amount}")
print(f"Your Discount is {discount}%")
print(f"Your Discount Amount is Rs. {amount*(discount/100)}")
print(f"Your Final Amount is Rs. {amount - (amount*(discount/100))}")
```
