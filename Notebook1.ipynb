```python
#list is also a type of data in Python.
'''Definition of a List: It is a collection of items which can be of different data types. It is ordered, 
changeable, and allows duplicate values. 
Lists are defined using square brackets [] and items are separated by commas.'''
```

```python
import numpy as np
all_types = [1, 3.14, "Hello", True, [1, 2, 3], {"name": "Alice", "age": 30}, (1, 2), {1, 2, 3}]

#print(type(all_types[-1]))
#all_types[5]["name"]

def tell_list_item_type(lst):
    for item in lst:
        print(f"{item} is of type {type(item)}")

tell_list_item_type(all_types)
```
